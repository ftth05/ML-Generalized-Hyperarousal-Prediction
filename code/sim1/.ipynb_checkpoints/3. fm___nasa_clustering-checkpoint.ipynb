```python
import pandas as pd
import numpy as np

import xlrd
import glob

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

from matplotlib.ticker import MaxNLocator


from sklearn.cluster import KMeans


project_path = '../../data/sim1/forecasting_markers/'
figure_path = '../../figure/sim1/forecasting_markers/'
```

```python
df = pd.read_csv(project_path + 'data_2.csv')
print(df.shape)
print(df.columns)

df.rename(columns={'Nasa_Total': 'NASA Total Sum'}, inplace=True)
```

```python
nasa_cols = ['NASA Total Sum', 'Effort', 'Frustration', 'Mental Demand', 'Performance', 'Physical Demand', 'Temporal Demand']
nasa_normalized_cols = [col + ' - Normalized' for col in nasa_cols]
nasa_all_cols = nasa_cols + nasa_normalized_cols
# nasa_df = df.groupby(['Subject', 'Drive', 'Drive_Label'])[nasa_cols].first().reset_index().sort_values(by=['Total Sum'])
nasa_df = df.groupby(['Subject', 'Drive', 'Drive_Label'])[nasa_all_cols].first().reset_index()
nasa_df
```

#  K-Means Clustering

#### The idea behind good clustering is having a small value of inertia, and small number of clusters.

- The value of inertia decreases as the number of clusters increase. So, its a trade-off here. 
- Rule of thumb: The elbow point in the inertia graph is a good choice because after that the change in the value of inertia isn’t significant.

```python
def get_intertia_plot(X):
    inertia_list = []
    
    for n in range(1, 10):
        model = KMeans(n_clusters=n)
        model.fit(X)
        inertia_list.append(model.inertia_)

    ######################################################################        
    fig = plt.figure(figsize=(14, 8))
    sns.set(font_scale=1.5)

    plt.figure(1, figsize = (12, 6))
    plt.plot(np.arange(1, len(inertia_list)+1), inertia_list, 'o')
    plt.plot(np.arange(1, len(inertia_list)+1), inertia_list, '-', alpha = 0.5)
    plt.xlabel('Total Clusters'), plt.ylabel('Inertia Value')

    plt.title(drive_label)

    plt.savefig(figure_path + 'nasa_' + col.replace(" ", "_").lower() + '_' + drive_label + '_inertia.png')
    plt.savefig(figure_path + 'nasa_' + col.replace(" ", "_").lower() + '_' + drive_label  + '_inertia.pdf')

    plt.show()
    ######################################################################
    
    
def get_clusters(X):
    model = KMeans(n_clusters=2)
    model.fit(X)  
    
    return model.labels_



def set_cluster_label_for_row(row, lo, med, hi):
    if row.Temp_Nasa_Cluster == lo:
        return 'Low'
    elif row.Temp_Nasa_Cluster == med:
        return 'Medium'
    elif row.Temp_Nasa_Cluster == hi:
        return 'High'

    
    

def set_cluster_labels(drive_df):
    drive_df['Temp_Nasa_Cluster'] = get_clusters(X)
    drive_df = drive_df.sort_values(by=['NASA Total Sum'])
    
    
    ##########################################################################################
    lo = drive_df.iloc[0]['Temp_Nasa_Cluster']
    hi = drive_df.iloc[-1]['Temp_Nasa_Cluster']
    med = list(set([0, 1, 2]) - set([lo, hi]))[0]

    drive_df['Nasa_Cluster'] = drive_df.apply(lambda row: set_cluster_label_for_row(row, lo, med, hi), axis=1)
    ##########################################################################################
    
    
#     for cluster in ['Low', 'Medium', 'High']:
#         temp_df = drive_df.copy()[drive_df.Nasa_Cluster==cluster]
#         print(cluster + ' -->')
#         print(sorted(temp_df.Subject.unique()))
    
    
    
    
    ##########################################################################################
    fig = plt.figure(figsize=(14, 8))
    ax = fig.add_subplot(111)

    ax.set_yticklabels([])
    ax.set_yticks([])

    for label in set(drive_df.Nasa_Cluster):
        cluster = list(drive_df[drive_df.Nasa_Cluster==label][col])
        plt.scatter(cluster, 
                    [1 for _ in range(len(cluster))],
                    label='Cluster {}'.format(label))

    plt.xlabel(col)
    plt.legend()
    
    plt.title(drive_label)

    plt.savefig(figure_path + 'nasa_' + col.replace(" ", "_").lower() + '-' + drive_label + '_cluster_v2.png')
    plt.savefig(figure_path + 'nasa_' + col.replace(" ", "_").lower() + '-' + drive_label + '_cluster_v2.pdf')
    
    plt.show()
    ##########################################################################################
    
    
    return drive_df
```

```python
nasa_cluster_df = pd.DataFrame()

for drive_label in sorted(nasa_df.Drive_Label.unique()):
# for drive_label in ['CD']:
    
    print('-----   ' + drive_label + '   -----')
    drive_df = nasa_df.copy()[nasa_df.Drive_Label==drive_label]
    
    # for col in nasa_cols:
    for col in ['NASA Total Sum']:
        
        X = drive_df[[col]].values
        
        # ValueError: n_samples=12 should be >= n_clusters=13
        if len(X) > 12:
            get_intertia_plot(X)
            drive_df = set_cluster_labels(drive_df)
            nasa_cluster_df = nasa_cluster_df.append(drive_df)
            
            
print(nasa_cluster_df.shape)
print(nasa_cluster_df.head())


# nasa_cluster_df.to_csv("../../data/sim1/forecasting_markers/nasa_cluster_df.csv", sep=',')
nasa_cluster_df.to_csv(project_path + 'nasa_cluster_df.csv', sep=',', index=False)
```

```python
for drive_label in nasa_cluster_df.Drive_Label.unique():
    print('\n------------------')
    print('-----   ' + drive_label + '   -----')
    print('------------------')
    temp_nasa_df = nasa_cluster_df.copy()[nasa_cluster_df.Drive_Label==drive_label]
        
    for cluster in ['Low', 'Medium', 'High']:
        temp_df = temp_nasa_df.copy()[temp_nasa_df.Nasa_Cluster==cluster]
        print(cluster + ' -->')
        print(sorted(temp_df.Subject.unique()))
        print('\n')
```

```python
get_max_limit = lambda col: 120 if col in ['NASA Total Sum', 'NASA Total Sum - Normalized'] else 20

def get_nasa_hist_plots(normalized=False):
    
    if normalized:
        cols = nasa_normalized_cols
        normalized_str = '_normalized'
    else:
        cols = nasa_cols
        normalized_str = ''
        
    
    for drive_label in nasa_cluster_df.Drive_Label.unique():
#     for drive_label in ['CD']:

        print('-----   ' + drive_label + '   -----')
        drive_df = nasa_cluster_df.copy()[nasa_cluster_df.Drive_Label==drive_label]

        fig, axes = plt.subplots(len(nasa_cols), 1, figsize=(15, 3*len(nasa_cols)), sharey=True, constrained_layout=True)
        fig.suptitle(drive_label)
        # fig.tight_layout()

        for i, col in enumerate(cols):
            nasa_values = drive_df[[col]].values

            ax=axes[i]     
            sns.histplot(data=drive_df,
                        x=col,
                        ax=ax,
                        palette="bright",
                        fill=True,
                        alpha = 0.7,
                        kde=True)

            if normalized:
                ax.set_xlim(min(nasa_values), get_max_limit(col))
            else:
                ax.set_xlim(1, get_max_limit(col))
                
            ax.set_title(col)
            ax.set_xlabel("")
            ax.xaxis.set_major_locator(MaxNLocator(integer=True))
            ax.axvline(drive_df[[col]].values.mean(), 0, 999, color = 'red', linestyle='--')
            
            
            lo_max = drive_df[drive_df.Nasa_Cluster=='Low'][col].max()
            hi_min = drive_df[drive_df.Nasa_Cluster=='High'][col].min()
            print(lo_max, hi_min)
            
            ## Remember the balck line indicates how clustering based on 'Total Sum' reflects on the subscals of NASA
            if not normalized and col in ['NASA Total Sum', 'NASA Total Sum - Normalized']:
                ax.axvline((lo_max+hi_min)//2, 0, 999, color = 'black', linestyle='--')

        plt.savefig(figure_path + 'nasa_hist_'+ drive_label.lower() + normalized_str + '.png')
        plt.savefig(figure_path + 'nasa_hist_' + drive_label.lower() + normalized_str + '.pdf')

        plt.show()     
    
```
